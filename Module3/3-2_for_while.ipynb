| 用法 | 說明 |
|:---:|:---:|
| `range(m)` | 走訪 0 到 m-1 的數字 |
| `range(n, m)` | 走訪 n 到 m-1 的數字  |
| `range(n, m, step)` | 以step為間距，走訪 n 到 m-1 的數字 |

##### for

```python
# for <變數> in <資料>:
for char in "哈嘍你好嗎":
    print(char, end=", ")  # 預設是給 \n -> 換行符號
```

```python
data = [1, 2, 3, 4, 5]
for num in data:
    if num > 3:  # 拿出來的資料大於 3 才要執行
        print(num)
```

```python
# range(m) -> 遍歷 0 到 m-1 的數字
for i in range(5):
    print("Hello World!")

for o in range(5):
    print(o)
```

```python
# range(n, m)

for i in range(1, 11):
    print(i)
```

```python
# range(n, m-1, step)
for i in range(2, 11, 2):
    print(i, end=", ")
```

### 小試身手 for  
請產出 a 開頭的 n 個偶數(含 a 本身)  
Ex:  
a = 5, n = 3  
Output : 6 8 10

a = 4, n = 5  
Output : 4 6 8 10 12

```python
a = -2
n = 20
for i in range(n):
    if (a % 2) == 0:
        print(a + 2 * i)
    else:
        print(a + 1 + 2 * i)
```

```python
# 嘗試產出
"""
* * * * *
 * * * *
  * * *
   * *
    *
"""
for i in range(5):
    for k in range(i):
        print(" ", end="")
    for j in range(5 - i):
        print("* ", end="")
    print(end="\n")


# 嘗試產出
"""
    *
   * *
  * * *
 * * * *
* * * * *
"""
for i in range(5):
    for j in range(5 - i - 1):
        print(" ", end="")
    for k in range(i + 1):
        print("* ", end="")
    print(end="\n")

#分隔線
print("="*50)

n = 8
for i in range(n):
    print(" " * i + "* " * (n - i - 1) + "*" * 1)

for i in range(n):
    print(" " * (n - i - 1) + "* " * i + "*" * 1)
```
