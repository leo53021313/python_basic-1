```python
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] # 純string是錯誤的寫法
    }
}
print(fruit_dict)
```

```python
# 使用 for 迴圈讀取 dict 資料1 
#只輸出key
for key in fruit_dict:
    print(key)
```

```python
# 使用 for 迴圈讀取 dict 資料2
#只輸出value
for value in fruit_dict.values():
    print(value)
```

```python
# 使用 for 迴圈讀取 dict 資料3
#key 和 value 都輸出
for key, value in fruit_dict.items():
    print(key)
    print(value)
```

```python
# 取得 dict 的 value
for key, value in fruit_dict.items():
    # print(value['color'])
    for color in value["color"]:
        print(color)
```

```python
#所有數量的加總
total = 0
for value in fruit_dict.values():
    print(value)
    total += value["quantity"]
print(f"總數量: {total}")
```

```python
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'}]

```

```python
# 將資料整理成 {'apple': ['red', 'green'], 'grape': ['purple', 'green'], 'kiwi': ['yellow', 'green']}

fruit_color_list2 = {}
for item in fruit_color_list:
    # print(item)
    for fruit, color in item.items():
        # print(fruit, color)
        if fruit not in fruit_color_list2:
            fruit_color_list2.update({fruit: [color]})
        elif fruit in fruit_color_list2:
            fruit_color_list2[fruit].append(color)

print(fruit_color_list2)

fruit_color_list2 = {}
for item in fruit_color_list:
    # print(item)
    for fruit, color in item.items():
        # print(fruit, color)
        if fruit_color_list2.get(fruit) is None:
            fruit_color_list2[fruit] = [color] #建立這個 key 裡面放 list[color]
        else:
            fruit_color_list2[fruit].append(color)

print(fruit_color_list2)
```

##### `defaultdict(<type>)`

```python
# from 模組 import 函式/class
from collections import defaultdict

# 當訪問的鍵不存在時，可以自動創建鍵和預設值，而不是拋出KeyError。
data = defaultdict(int)
print(data)
print(type(data))

# 以int型態的預設值 0 來初始化
print(data["apple"])
print(data)
```

```python
result = defaultdict(list) # 預設值為 [] list
print(result["Apple"])
print(result)
```

```python
# 資料處理
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'},
    {'apple': 'red'}
]

result = defaultdict(list) # 預設值為 [] list
# print(result["Apple"])
# print(result)
for fruit_color in fruit_color_list:
    for fruit, color in fruit_color.items():
        if color not in result[fruit]: # 如果顏色不再 list 中才 .append()
            result[fruit].append(color) #呼叫到該 key 的同時 value 創建一個 []，所以可以直接 append color
print(result)
        
```

```python
from collections import defaultdict
fruit_color_list = [
    {'apple': 'red'}, 
    {'grape': 'purple'}, 
    {'kiwi': 'yellow'}, 
    {'apple': 'green'}, 
    {'kiwi': 'green'}, 
    {'grape': 'green'},
    {'apple': 'red'}
]

result = {}
# {}
# {"": ""} 初始化 value 會是 空白string
result1 = defaultdict(list)
# {"": []} 初始化 value 會是 空白list
# {key: [value]}
# result.update({key: [value]}) 一樣的東西
print(result)
print(result1)

for fruit in fruit_color_list:
    for key, value in fruit.items():
        if result.get(key) is None:             # key不存在result裡面時新增
            # result = {key: value}             # = 是錯誤寫法，會覆蓋原本的result內容
            result.update({key: [value]})       # 要用.update() 讓 result新增新的dictionary內容，而不會覆蓋result原有的內容
        elif result.get(key) is not None:       # key存在result裡面時，只更新他的value(list)
            if value not in result[key]:        # 檢查每個水果的顏色有沒有重複
                result[key].append(value)       # 沒有重複才新增水果的顏色

print(result)
```
