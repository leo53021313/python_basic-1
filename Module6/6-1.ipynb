```python
# 類別名稱
# 類別屬性 (Class Attribute)
class Car: # 定義一個 類別(class)
    wheels = 4 # 定義 Car 的 類別屬性(class Attribute)
    # 呼叫Car.wheels 得到 4

print(Car)
print(Car.wheels)
```

```python
# 基於需要哪些資料或資訊來使用底下的 method 才建立 class
```

```python
# 類別名稱
class Car:
    # 類別屬性 (Class Attribute)
    wheels = 4

    #建構函式
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model
        # 創建Car類別的一個實例: car1 = Car("Toyota", "Camry")
        # 使用Car的實例屬性: car1.brand 得到 Toyota

    # 方法 (Method)
    def show_info(self, year=2026):
        print(f"{year} {self.brand} has model - {self.model} and has {self.wheels} wheels.")

# 建立實例
car1 = Car("Mercedes", "G63")
print(car1.brand)
print(car1.model)
car1.show_info()

car2 = Car("BMW", "M5")
car2.show_info(year = 1999)
```
