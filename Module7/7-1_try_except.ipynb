##### try...except...

```python
# 可以連續定義數個例外處理機制
# 最符合的例外事件類別會先拋出

try:
    x = 100
    print(y) # 發生錯誤 就跳到 except 繼續執行
    print(x)
except:
    print("發生錯誤")
```

```python
try:
    x = 100
    print(y) # 發生錯誤 就跳到 except 繼續執行
    print(x)

#as -> 左邊赴值給右邊的
except Exception as e: #except 接錯誤類別
    print(f"發生錯誤: {e}")
```

```python
# 多層例外處理:
try:
    x = 100
    print(y) # 發生錯誤 就跳到 except 繼續執行
    print(x)

except NameError as e: #指定錯誤類別
    print(f"發生 NameError: {e}")
except Exception as e: #except 接錯誤類別
    print(f"發生錯誤: {e}")

#Exception永遠寫在最下面(因為他已經囊括了90% error)，上面給其他特定的錯誤

```

##### try...except...else...finally

```python
# 如果沒有錯誤，就會執行 else 程式區塊
# finally 程式區塊是無論有無錯誤，都會執行

# 我的程式報錯了，該如何確保資料庫連線有中斷


try:
    x = 10
    # z = 5
    print(z)
except Exception as e:
    print(f"程式報錯了: {e}")
else:
    print("沒有錯誤發生，進到 else")
finally:
    print(f"無論如何都要執行: x = {x}")
```
