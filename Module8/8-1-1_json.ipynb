### `json` 套件

| 用法 | 說明 |
|:--:|:--:|
| .dump() |物件轉JSON並寫入文件|
| .dumps() |物件轉JSON字串|
| .load() |讀取JSON文件|
| .loads() |將JSON字串轉成Python物件|

```python
# 引入 json
import json
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange'] # 純string是錯誤的寫法
    }
}

# json 格式 -> 在Python只能是 list, dict
with open("fruit.json", "w") as f:
    f.write(json.dumps(fruit_dict)) # json.dumps() 將物件轉 json 字串
    
```

```python
import json
fruit_dict = {    
    'apple': {
        'price': 100,
        'quantity': 50,
        'color': ['green', 'red']
    },
    'orange': {
        'price': 50,
        'quantity': 30,
        'color': ['orange']
    }
}

with open("fruit2.json", "w") as f:
    json.dump(fruit_dict, f, indent=4) # json.dump(<list, dict>, <file>) 將物件轉JSON並寫入文件
    
```

```python
# loads()
with open("fruit2.json", "r") as f:
    data = f.read()

print(type(data))
print(data)
fruit_dict = json.loads(data)  # 讀取字串轉物件
print(type(fruit_dict))
print(fruit_dict)
```

```python
# load()
with open("fruit2.json", "r") as f:
    data = json.load(f)

print(type(data))
print(data)
```
