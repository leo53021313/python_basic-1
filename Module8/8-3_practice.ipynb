##### 將 data.txt 的資料分類成另外三個檔案，分別為 cc.txt 、 css.txt 、 c.txt  

```python
import os
file = "data.txt"
out_file1 = "cc.txt"
out_file2 = "css.txt"
out_file3 = "c.txt"
with open(file, "r", encoding="utf-8") as f:
    items = f.readlines()
    for item in items:
        str_list = item.split(".")
        if str_list[1].strip("\n") == "cc":
            with open(out_file1, "a", encoding="utf-8") as f1:
                f1.write(item)
        elif str_list[1].strip("\n") == "css":
            with open(out_file2, "a", encoding="utf-8") as f2:
                f2.write(item)
        elif str_list[1].strip("\n") == "c":
            with open(out_file3, "a", encoding="utf-8") as f3:
                f3.write(item)

print(items)
```

```python
import os
class file_process:
    input_file = ""

    def __init__(self, input_file="data.txt"):
        self.input_file = input_file

    def read_file(self):
        with open(self.input_file, "r", encoding="utf-8") as f:
            return f.readlines()
        
    def write_file(self, file_path, content):
        with open(file_path, "a", encoding="utf-8") as f:
            f.write(content)

    def clear_file(self):
        if os.path.exists("cc.txt"):
            os.remove("cc.txt")
        if os.path.exists("css.txt"):
            os.remove("css.txt")
        if os.path.exists("c.txt"):
            os.remove("c.txt")

fp = file_process()
fp.clear_file()
for item in fp.read_file():
    file_extension = item.strip("\n").split(".")[1]
    if file_extension == "cc":
        fp.write_file("cc.txt", item)
    elif file_extension == "css":
        fp.write_file("css.txt", item)
    elif file_extension == "c":
        fp.write_file("c.txt", item)
```
