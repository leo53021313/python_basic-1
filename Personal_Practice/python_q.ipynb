### 有一筆學生資料如下:

```python
students_grade = [
    {"name": "Denji", "國文": 30, "英文": 10, "數學": 20},
    {"name": "Power", "國文": 60, "英文": 85, "數學": 70},
    {"name": "Aki", "國文": 95, "英文": 88, "數學": 92},
    {"name": "Pochita", "國文": 50, "英文": 65, "數學": 20},
    {"name": "Makima", "國文": 100, "英文": 98, "數學": 95},
    {"name": "Reze", "國文": 65, "英文": 40, "數學": 30},
]
```

##### 1. 印出所有學生姓名

```python
# 0123456789..... index 適合list
# list: [a, b, c, d]
#  print(list[0])
# dict: {"name":Leo, "age":18}


for grade in students_grade:
    print(grade["name"])
```

##### 2. 印出 `Pochita` 的數學成績

```python
for grade in students_grade:
    if grade["name"] == "Pochita":
        print(grade["數學"])
```

#### 3. 計算每位學生的總分
輸出格式  
`Phochita 總分: xxx 分`

```python
for grade in students_grade:
    total = grade["國文"] + grade["英文"] + grade["數學"]
    print(f"{grade["name"]} 總分: {total} 分")
    
```

#### 4. 計算每位學生的平均分數
輸出格式  
`Pochita 平均分數: xxx 分`

```python
for grade in students_grade:
    total = grade["國文"] + grade["英文"] + grade["數學"]
    print(f"{grade["name"]} 平均分數: {total/3:.2f} 分")

# float
```

#### 5. 找出數學最高分的學生

```python
highest_score = 0
highest_student = "" 
for grade in students_grade:
     if grade["數學"] > highest_score:
          highest_score = grade["數學"]
          highest_student = grade["name"]

print(highest_student)

# for grade in students_grade:
#      if grade["數學"] == highest_score:
#           print(grade["name"])
```

#### 6. 計算全班國文平均

```python
chinese_total = 0
for grade in students_grade:
    chinese_total += grade["國文"]

# print(chinese_total)
# print(len(students_grade))
# length
print(chinese_total/len(students_grade))
```

#### 7. 找出不及格的學生（任一科 < 60）

```python
for grade in students_grade:
    if grade["國文"] < 60 or grade["英文"] < 60 or  grade["數學"] < 60:
        print(grade["name"])
```

#### 8. 新增「總分」到每個 dict
例如 :  
`{"name": "Pochita", "國文": 80, ..., "總分": 245}`

```python
for student in students_grade:
    student["總分"] = student["國文"] + student["英文"] + student["數學"]

print(students_grade)
```

#### 9. 找出三科都 > 80 的學生

```python
for grade in students_grade:
    if grade["國文"] > 80 and grade["英文"] > 80 and  grade["數學"] > 80:
        print(grade["name"])
```

#### 10. 將資料轉成以下格式 (dict of dict)

```python
"""
{
    "Denji": {"國文": 30, "英文": 10, "數學": 20},
    .....
}
"""

[
    {"name": "Denji", "國文": 30, "英文": 10, "數學": 20},
    {"name": "Power", "國文": 60, "英文": 85, "數學": 70},
    {"name": "Aki", "國文": 95, "英文": 88, "數學": 92},
    {"name": "Pochita", "國文": 50, "英文": 65, "數學": 20},
    {"name": "Makima", "國文": 100, "英文": 98, "數學": 95},
    {"name": "Reze", "國文": 65, "英文": 40, "數學": 30},
]
# list + 字典
# 字典 key = 名字 值(vaule) = 字典 成績

students_grade_dict = {}
for student in students_grade:
    # {"name": "Denji", "國文": 30, "英文": 10, "數學": 20}
    students_grade_dict[student["name"]] = {
            "國文":student["國文"],
            "英文":student["英文"],
            "數學":student["數學"]
    }

print(students_grade_dict)
```
